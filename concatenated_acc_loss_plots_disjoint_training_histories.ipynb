```python
import re
import matplotlib.pyplot as plt
```

```python
training_history_path = "epochs_1_41_concat.txt"
```

```python
with open(training_history_path, 'r') as file:
    txt = file.read()
```

```python
acc = re.findall(r' accuracy: [0]+([.][0-9]*)', txt)
val_acc = re.findall(r' val_accuracy: [0]+([.][0-9]*)', txt)
loss = re.findall(r' loss: [0]+([.][0-9]*)', txt)
val_loss = re.findall(r' val_loss: [0]+([.][0-9]*)', txt)
```

```python
acc = list(map(float, acc))
val_acc = list(map(float, val_acc))
loss = list(map(float, loss))
val_loss = list(map(float, val_loss))
```

```python
len(acc), len(val_acc), len(loss), len(val_loss)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(acc)
plt.plot(val_acc)
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.subplot(1,2,2)
plt.plot(loss)
plt.plot(val_loss)
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```
